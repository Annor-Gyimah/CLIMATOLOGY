```python
import xarray as xr
import matplotlib.pyplot as plt

# define file path
file_path = 'CHILE_RR.nc' # replace with your own file path

# read in data as an xarray Dataset
ds = xr.open_dataset(file_path)

# select region of interest
lon_min, lon_max = -109.5, -66.25
lat_min, lat_max = -55.75, -17.0
ds_roi = ds.sel(lon=slice(lon_min, lon_max), lat=slice(lat_min, lat_max))
ds_roi = ds_roi.sel(datetime=slice('2010-01','2019-12'))
# calculate annual totals and average
annual_totals = ds_roi.precip.resample(datetime='1AS').sum(dim='datetime')
annual_average = ds_roi.precip.resample(datetime='1AS').mean(dim='datetime')

# create time-series plot of annual totals and average
fig, ax = plt.subplots()
annual_totals.plot(ax=ax, label='Annual Totals')
annual_average.plot(ax=ax, label='Annual Average')
ax.set(xlabel='Year', ylabel='Rainfall (mm)', title='Rainfall Totals and Average per Year')
ax.legend()

plt.show()
```
